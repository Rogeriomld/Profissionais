```python
import pandas as pd

# Carrega todas as linhas do arquivo como texto
with open('Relatorio_Atendimento_dos_Profissionais_janeiro.csv', 'r', encoding='latin1') as f:
    linhas = f.readlines()

blocos = []
nome = ''
cargo = ''
quantidade1 = ''
quantidade2 = ''
tempo_total = ''

for linha in linhas:
    linha = linha.strip()

    # Detectar linhas com nome + cargo + quantidade
    if ';' in linha and linha.count(';') >= 5 and not linha.split(';')[0].isdigit():
        partes = linha.split(';')
        nome = partes[0].strip()
        cargo = partes[1].strip()
        quantidade1 = partes[2].strip()
        quantidade2 = partes[3].strip()
        tempo_total = partes[4].strip()

        # Armazena esse registro de quantidade
        blocos.append({
            'Nome': nome,
            'Cargo': cargo,
            'Quantidade1': quantidade1,
            'Quantidade2': quantidade2,
            'Tempo': tempo_total,
            'Setor': '',
            'Tipo_Atendimento': '',
            'Data_Entrada': '',
            'Data_Saida': '',
            'Duracao': ''
        })

    # Linha de atendimento (início com código)
    elif linha and linha[0].isdigit() and ';' in linha:
        partes = linha.split(';')
        bloco = {
            'Nome': nome,
            'Cargo': cargo,
            'Quantidade1': '',
            'Quantidade2': '',
            'Tempo': '',
            'Setor': partes[3].strip() if len(partes) > 3 else '',
            'Tipo_Atendimento': partes[4].strip() if len(partes) > 4 else '',
            'Data_Entrada': partes[6].strip() if len(partes) > 6 else '',
            'Data_Saida': partes[7].strip() if len(partes) > 7 else '',
            'Duracao': partes[8].strip() if len(partes) > 8 else ''
        }
        blocos.append(bloco)

# Criar DataFrame
df_tudo = pd.DataFrame(blocos)

# Salvar como CSV
df_tudo.to_csv('base_completa_com_quantidades_e_atendimentos.csv', sep=';', index=False, encoding='latin1')
```
